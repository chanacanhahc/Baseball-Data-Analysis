```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("2019_kbo_for_kaggle_v2.csv")
df
```

```python
df['year']
```

```python
#require-1
d=df[['batter_name','H','avg','HR','OBP','year']]
for i in range(2015,2019):
    a=d.loc[d.year==i]
    H,avg,HR,OBP=a[['batter_name','H']],a[['batter_name','avg']],a[['batter_name','HR']],a[['batter_name','OBP']]
    H=H.sort_values(ascending=False,by=['H']).head(10)
    avg=avg.sort_values(ascending=False,by=['avg']).head(10)
    HR=HR.sort_values(ascending=False,by=['HR']).head(10)
    OBP=OBP.sort_values(ascending=False,by=['OBP']).head(10)
    print(f"{i}년 탑10 선수 이름")
    print("안타:",*H.values[:,0])
    print("타율:",*avg.values[:,0])
    print("홈런:",*HR.values[:,0])
    print("출루율:",*OBP.values[:,0])
```

```python
d=df.loc[df.year==2018][['batter_name','war','cp']]
C=d.loc[d.cp=="포수"].sort_values(ascending=False,by=['war']).head(1)
FB=d.loc[d.cp=="1루수"].sort_values(ascending=False,by=['war']).head(1)
SB=d.loc[d.cp=="2루수"].sort_values(ascending=False,by=['war']).head(1)
TB=d.loc[d.cp=="3루수"].sort_values(ascending=False,by=['war']).head(1)
S=d.loc[d.cp=="유격수"].sort_values(ascending=False,by=['war']).head(1)
LF=d.loc[d.cp=="좌익수"].sort_values(ascending=False,by=['war']).head(1)
CF=d.loc[d.cp=="중견수"].sort_values(ascending=False,by=['war']).head(1)
RF=d.loc[d.cp=="우익수"].sort_values(ascending=False,by=['war']).head(1)
print("2018년 각 포지션 별 승리 기여도가 제일 높은 선수")
print(*C.values[0])
print(*FB.values[0])
print(*SB.values[0])
print(*TB.values[0])
print(*S.values[0])
print(*LF.values[0])
print(*CF.values[0])
print(*RF.values[0])
```
